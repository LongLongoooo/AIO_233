Mean

```python
import numpy as np

def compute_mean(X):
  return np.sum(X)/len(X)
```

```python
X = np.array([2, 0, 2, 2, 7, 4, -2, 5, -1, -1])
compute_mean(X)
```

Median

```python
def compute_median(X):
  size = len(X)
  X = np.sort(X)
  print(X)
  if size % 2 == 0:
    i = int(size/2)
    return (X[i] + X[i-1])/2

  else:
    return X[int((size - 1)/2)]
```

```python
X =np.array( [1, 5, 4, 4, 9, 13] )
print (" Median : ", compute_median (X))
```

Standard deviation

```python
def compute_std(X):
  mean = compute_mean(X)
  mean_pow = compute_mean(X**2)
  variance = mean_pow - mean**2
  return np.sqrt(variance)
```

```python
X = np.array([ 171 , 176 , 155 , 167 , 169 , 182])
print (compute_std (X))
```

Correlation cofficient


```python
def compute_correlation_cofficient(X, Y):
  mean_x = compute_mean(X)
  mean_y = compute_mean(Y)
  numerator = compute_mean((X - mean_x)*(Y - mean_y))

  var_x = compute_std(X)
  var_y = compute_std(Y)
  denominator = var_x * var_y

  return np.round(numerator/denominator, 2)
```

```python
X = np. array ([-2, -5, -11, 6, 4, 15, 9])
Y = np. array ([4 , 25, 121 , 36, 16, 225 , 81])
print (" Correlation : ", compute_correlation_cofficient (X,Y))
```

Cosine Similarity

```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

def compute_cosine_similarity(matrix1, matrix2):
    """
    Compute cosine similarity between two matrices.
    """
    dot_product = np.dot(matrix1, matrix2.T)
    len_matrix1 = np.linalg.norm(matrix1, axis=1, keepdims=True)
    len_matrix2 = np.linalg.norm(matrix2, axis=1, keepdims=True).T

    return dot_product / (len_matrix1 * len_matrix2)

def tfidf_search(question, tfidf_vectorizer=TfidfVectorizer(), top_d=5):
    """
    Search the top k documents similar to the query using TF-IDF and cosine similarity.
    """
    # Define data and convert it into array
    vi_data_df = pd.read_csv("/content/vi_text_retrieval.csv")
    context = vi_data_df["text"]
    context = [doc.lower() for doc in context]

    # Fit and transform the context using TF-IDF
    context_embedded = tfidf_vectorizer.fit_transform(context).toarray()

    # Lowercase the query and transform using the same TF-IDF vectorizer
    query_embedded = tfidf_vectorizer.transform([question.lower()]).toarray()

    # Compute cosine similarity for all documents at once
    cosine_scores = compute_cosine_similarity(context_embedded, query_embedded)

    # Flatten the cosine_scores array to 1D
    cosine_scores = cosine_scores.flatten()

    # Get top k cosine score and index its
    result = []
    for idx in cosine_scores.argsort()[-top_d:][::-1]:
        doc_score = {
            "id": idx,
            "cosine_score": cosine_scores[idx]
        }
        result.append(doc_score)

    return result
```

```python
question = vi_data_df.iloc[0]["question"]
print(question)
results = tfidf_search ( question, tfidf_vectorizer , top_d =5)
results [0][ "cosine_score"]
```

```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

def compute_correlation(vector1, vector2):
    """
    Compute the Pearson correlation coefficient between two vectors.
    """
    mean_vector1 = np.mean(vector1, axis=1, keepdims=True)
    mean_vector2 = np.mean(vector2, axis=1, keepdims=True)

    norm_vector1 = vector1 - mean_vector1
    norm_vector2 = vector2 - mean_vector2

    dot_product = np.dot(norm_vector1, norm_vector2.T)
    norm1 = np.linalg.norm(norm_vector1, axis=1, keepdims=True)
    norm2 = np.linalg.norm(norm_vector2, axis=1, keepdims=True).T

    return dot_product / (norm1 * norm2)

def corr_search(question, tfidf_vectorizer, top_d=5):
    """
    Search the top k documents similar to the query using TF-IDF and Pearson correlation.
    """
    # Define data and convert it into array
    vi_data_df = pd.read_csv("/content/vi_text_retrieval.csv")
    context = vi_data_df["text"]
    context = [doc.lower() for doc in context]

    # Fit and transform the context using TF-IDF
    context_embedded = tfidf_vectorizer.fit_transform(context).toarray()

    # Lowercase the query and transform using the same TF-IDF vectorizer
    query_embedded = tfidf_vectorizer.transform([question.lower()]).toarray()

    # Compute correlation scores for all documents at once
    corr_scores = compute_correlation(context_embedded, query_embedded)

    # Flatten the corr_scores array to 1D and exclude the query itself
    corr_scores = corr_scores.flatten()
    corr_scores = corr_scores[1:]

    # Get top k correlation score and index its
    results = []
    for idx in corr_scores.argsort()[-top_d:][::-1]:
        doc = {
            'id': idx,
            'corr_score': corr_scores[idx]
        }
        results.append(doc)

    return results
```

```python
question = vi_data_df.iloc [0]["question"]
results = corr_search( question, tfidf_vectorizer, top_d =5)
print(results)
results[0]["corr_score"]
```
